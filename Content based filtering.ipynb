```
import pandas as pd
movies = pd.read_csv('movies.csv',error_bad_lines=False)
```

```
movies
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
movies["overview"] = movies["overview"].fillna("")
```

```
tfidf_matrix = tfidf.fit_transform(movies["overview"])
```

```
pd.DataFrame(tfidf_matrix.todense(), columns=tfidf.get_feature_names())
```

```
tfidf_matrix.shape
```

### Similarity Matrix

```
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```
similarity_matrix
```

### Find the most similar movies to a certain movie

```
movies_title = "The Dark Knight Rises"
idx = movies[movies["title"] == movies_title].index[0]
idx
```

```
score = list(enumerate(similarity_matrix[idx]))
score
```

```
scores=sorted(score,key=lambda x:x[1],reverse=True)
scores
```

```
def similar_movies(movies_title,nr_movies):
    movies_index = movies[movies["title"] == movies_title].index[0]
    movie_list = sorted(list(enumerate(similarity_matrix[movies_index])),reverse=True, key=lambda x:x[1])
    movies_indices = [tpl[0] for tpl in movie_list[1:nr_movies+1]]
    similar_movies = list(movies['title'].iloc[movies_indices])
    return similar_movies
```

```
 similar_movies('Shutter Island',2)
```
